```python
import requests # to make TMDB API calls
import locale 
locale.setlocale( locale.LC_ALL, '' )
from time import sleep
from random import randint
import pandas as pd
import csv

api_key = "21aefd57526da8bc838461554ca6cc49" #TMDB API key

pages = [str(i) for i in range(1,100)]

columns = ['title','Release_date','Overview','Adult','Vote_count','Vote_average','Popularity']

with open('Test23.csv', 'w') as f:
        f.write(','.join(columns) + '\n')

for page in pages:
    response = requests.get('https://api.themoviedb.org/3/discover/movie?api_key=' +  api_key + '&primary_release_year=2002' + '&page=' + page) #URL fetched from API 

    movies = response.json() #Results appear as JSON object 

    movies_list = movies['results'] #convert the JSON object to list 

    df = pd.DataFrame(columns=columns)
        
    for film in movies_list: #store results in a dataframe 
        df.loc[len(df)]=[film["title"],film["release_date"],film["overview"],film["adult"],film["vote_count"],film["vote_average"],film["popularity"]] 
        
    with open('Test23.csv', 'a') as f: #export the results to a CSV file 
        export_csv = df.to_csv(f, index = None, header=False)
```

```python
results = api.GetUserTimeline(screen_name='CallMeKevin1811', count=100) #Storing CallMeKevin1811 Twitter Data
posts =[]
favourites =[]
time = []
user_id = []
name = []
followers = []
hashtags = []
for each in results:
    status = json.loads(str(each))
    posts.append(status['text'])
    try:
        favourites.append(status['favorite_count'])
    except:
        favourites.append(status['retweeted_status']['favorite_count'])
    time.append(status['created_at'])
    user_id.append(status['user']['id'])
    followers.append(status['user']['followers_count'])
    name.append(status['user']['name'])
    try:
        hashtags.append(status['hashtags'][0]['text'])
    except:
        hashtags.append(status['hashtags'])
```

```python
df_1 = pd.DataFrame(data = name, columns = ['Name'])
df_1['User_id'] = user_id
df_1['Followers'] = followers
df_1['Post'] = posts
df_1['Hashtags'] = hashtags
df_1['Posted_time'] = time
df_1['Favourites'] = favourites
df_1
```

```python
with open('Twitter_consumer.csv', 'a') as f:
    s1 = df_1.to_csv(f, index = None, header=True)
```
